```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cns=pd.read_csv('/content/conjunto_de_datos_iter_00CSV20.csv')
```

```python
mx=gpd.read_file('/content/00mun.shp')
```

```python

cvm=[]
for c in cns.MUN:
  if len(str(c))<3:
    if len(str(c))==2:
      cvm.append('0'+str(c))
    elif len(str(c))==1:
      cvm.append('00'+str(c))
  else:
    cvm.append(str(c))
```

```python
cve=[]
for c in cns.ENTIDAD:
  if len(str(c))<2:
    if len(str(c))==1:
      cve.append('0'+str(c))
  else:
    cve.append(str(c))
```

```python
CVEGEO=[]
for i in cns.index:
  c=cns.MUN[i]
  v=cns.ENTIDAD[i]
  if len(str(c))>=3:
    if len(str(v))>=2:
      CVEGEO.append(str(v)+str(c))
    elif len(str(v))==1:
      CVEGEO.append('0'+str(v)+str(c))
  elif len(str(c))==2:
    if len(str(v))>=2:
      CVEGEO.append(str(v)+'0'+str(c))
    elif len(str(v))==1:
      CVEGEO.append('0'+str(v)+'0'+str(c))
  else:
    if len(str(v))>=2:
      CVEGEO.append(str(v)+'00'+str(c))
    elif len(str(v))==1:
      CVEGEO.append('0'+str(v)+'00'+str(c))

```

```python
cns['CVEGEO']=CVEGEO
```

```python
D=mx.merge(cns,on='CVEGEO')
```
