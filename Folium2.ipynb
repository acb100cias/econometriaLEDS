```python
m2=folium.Map(location=(30,10),zoom_start=3,
           tiles='cartodb_positron')
```

```python
m2
```

```python
paises_url=('http://geojson.xyz/naturalearth-3.3.0/ne_50m_admin_0_countries.geojson')
```

```python
folium.GeoJson(paises_url).add_to(m2)
```

```python
m2
```

```python
eco=pd.read_excel('/content/NFBA 2023 Public Data Package 1.0.xlsx',
                  sheet_name='Country Results (2019)',skiprows=22,
                  skipfooter=18)
```

```python
eco
```

```python
eco=eco.fillna(0)
```

```python
eco
```

```python
eco.keys()
```

```python
folium.Choropleth(
  geo_data=paises_url,
  data=eco,
  columns=['Country','Total Ecological Footprint (Production)'],
   key_on='feature.properties.name',
  fill_color='RdYlGn_r',
   fill_opacity=0.8,
  line_opacity=0.3,
  nan_fill_color='white',
legend_name='Ecological footprint per capita'
).add_to(m2)
```

```python
m2
```

```python
folium.Choropleth(
  geo_data=paises_url,
  data=eco,
  columns=['Country','Total Ecological Footprint (Consumption)'],
   key_on='feature.properties.name',
  fill_color='RdYlGn_r',
   fill_opacity=0.8,
  line_opacity=0.3,
  nan_fill_color='white',
legend_name='Ecological footprint per capita'
).add_to(m2)
```
