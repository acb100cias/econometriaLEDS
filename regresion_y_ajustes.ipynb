```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x=30*np.random.random(size=(200,1))
```

```python
x
```

```python
y=0.5*x+1+np.random.normal(size=x.shape)
```

```python
y
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x,y)
```

```python
modelo=LinearRegression()
```

```python
modelo.fit(x,y)
```

```python
X=np.linspace(0,30,100)
Y=modelo.predict(X[:,np.newaxis])
```

```python
plt.figure(figsize=(5,5))
ax=plt.axes()
ax.scatter(x,y)
ax.plot(X,Y,'r')
```

```python
x=30*np.random.exponential(size=(200,1),scale=0.04)
y=0.5*x+1+np.random.exponential(size=x.shape,scale=4)
modelo=LinearRegression()
modelo.fit(x,y)
X=np.linspace(0,np.floor(x.max()),100)
Y=modelo.predict(X[:,np.newaxis])
plt.figure(figsize=(5,5))
ax=plt.axes()
ax.scatter(x,y)
ax.plot(X,Y,'r')
ax.plot(X,0.5*X+1,'g')
```

```python
import scipy.optimize as opt
```

```python
def f(x,a,b,c):
  return a*np.exp(-b*x)+c
x=np.linspace(0,4,100)
y=f(x,2.5,1.3,0.5)+0.2*np.random.normal(size=x.shape)
fx,fy=opt.curve_fit(f,x,y)
```

```python
plt.figure(figsize=(5,5))
ax=plt.axes()
ax.plot(x,y,'o')
ax.plot(x,f(x,*fx),'r')
```
