```python
mx=gpd.read_file('/content/00mun.shp')
```

```python
mx
```

```python
cvm=[]
for c in cn.MUN:
  if len(str(c))<3:
    if len(str(c))==2:
      cvm.append('0'+str(c))
    elif len(str(c))==1:
      cvm.append('00'+str(c))
  else:
    cvm.append(c)
```

```python
cvm
```

```python
ce=[]
for c in cn.ENTIDAD:
  if len(str(c))<2:
    if len(str(c))==1:
      ce.append('0'+str(c))
    else:
      ce.append(str(c))
```

```python
CVEGEO=[]
for i in cn.index:
  c=cn.MUN[i]
  v=cn.ENTIDAD[i]
  if len(str(c))>=3:
    if len(str(v))>=2:
      CVEGEO.append(str(v)+str(c))
    elif len(str(v))==1:
      CVEGEO.append('0'+str(v)+str(c))
  elif len(str(c))==2:
    if len(str(v))>=2:
      CVEGEO.append(str(v)+'0'+str(c))
    elif len(str(v))==1:
      CVEGEO.append('0'+str(v)+'0'+str(c))
  else:
    if len(str(v))>=2:
      CVEGEO.append(str(v)+'00'+str(c))
    elif len(str(v))==1:
      CVEGEO.append('0'+str(v)+'00'+str(c))
```

```python
cn['CVEGEO']=CVEGEO
```

```python
datos['CVEGEO']=datos.ubica_geo.astype('str')
```

```python
M=mx.merge(datos,on='CVEGEO')
```
